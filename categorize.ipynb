```python
import os
import glob
import pandas as pd
import shutil

# Create directories for organization
parquet_dir = "parquet_files"
csv_dir = "csv_files"

os.makedirs(parquet_dir, exist_ok=True)
os.makedirs(csv_dir, exist_ok=True)

# Find all parquet files in the current directory
parquet_files = glob.glob("*.parquet")
print(f"Found {len(parquet_files)} parquet files:")
for file in parquet_files:
    print(f"  - {file}")

# List to store all dataframes
all_dataframes = []

# Process each parquet file
for parquet_file in parquet_files:
    print(f"\nProcessing: {parquet_file}")
    
    # Read the parquet file
    df = pd.read_parquet(parquet_file)
    print(f"  Shape: {df.shape}")
    
    # Convert to CSV and save in csv_files directory
    csv_filename = parquet_file.replace('.parquet', '.csv')
    csv_path = os.path.join(csv_dir, csv_filename)
    df.to_csv(csv_path, index=False)
    print(f"  Saved CSV as: {csv_path}")
    
    # Move parquet file to parquet_files directory
    parquet_path = os.path.join(parquet_dir, parquet_file)
    shutil.move(parquet_file, parquet_path)
    print(f"  Moved parquet to: {parquet_path}")
    
    # Add to list for combining
    all_dataframes.append(df)

print(f"\n{'='*50}")
print("SUMMARY:")
print(f"Individual CSV files created in '{csv_dir}/' directory: {len(parquet_files)}")
print(f"Parquet files moved to '{parquet_dir}/' directory: {len(parquet_files)}")
```

```python
# Combine all dataframes into one
if all_dataframes:
    combined_df = pd.concat(all_dataframes, ignore_index=True)
    
    print(f"Combined dataset shape: {combined_df.shape}")
    print(f"Total rows: {combined_df.shape[0]:,}")
    print(f"Total columns: {combined_df.shape[1]}")
    
    # Save combined data to CSV in the csv_files directory
    combined_csv_filename = "all_data_combined.csv"
    combined_csv_path = os.path.join(csv_dir, combined_csv_filename)
    combined_df.to_csv(combined_csv_path, index=False)
    print(f"\nCombined data saved as: {combined_csv_path}")
    
    # Display basic info about the combined dataset
    print(f"\nColumns in combined dataset:")
    for i, col in enumerate(combined_df.columns, 1):
        print(f"  {i}. {col}")
    
    print(f"\nFirst few rows of combined data:")
    print(combined_df.head(3))
    
    # Check for any data quality issues
    print(f"\nData Quality Check:")
    print(f"  Missing values per column:")
    missing_values = combined_df.isnull().sum()
    for col, missing in missing_values.items():
        if missing > 0:
            print(f"    {col}: {missing} missing values")
    
    if missing_values.sum() == 0:
        print("    No missing values found!")
        
    # Final directory structure summary
    print(f"\n{'='*60}")
    print("FINAL ORGANIZATION:")
    print(f"📁 {parquet_dir}/ - Contains all original parquet files")
    print(f"📁 {csv_dir}/ - Contains all CSV files (individual + combined)")
    print(f"   └── Individual CSV files: {len(parquet_files)}")
    print(f"   └── Combined CSV file: {combined_csv_filename}")
else:
    print("No dataframes to combine!")
```

```python
import pandas as pd
import numpy as np

# Read both dataframes
print("Reading data files...")
fatture_df = pd.read_csv('fatture.csv')
categorizzazione_df = pd.read_csv('categorizzazione_fornitori.csv')

print(f"Fatture dataset shape: {fatture_df.shape}")
print(f"Categorizzazione dataset shape: {categorizzazione_df.shape}")

# Display column names to verify the key columns
print(f"\nFatture columns: {list(fatture_df.columns)}")
print(f"Categorizzazione columns: {list(categorizzazione_df.columns)}")

# Check the key columns
print(f"\nUnique values in fatture.RAGIONE_SOCIALE: {fatture_df['RAGIONE_SOCIALE'].nunique()}")
print(f"Unique values in categorizzazione.DENOMINAZIONE: {categorizzazione_df['DENOMINAZIONE'].nunique()}")

# Display sample values from both key columns
print(f"\nSample RAGIONE_SOCIALE values:")
print(fatture_df['RAGIONE_SOCIALE'].head(10).tolist())

print(f"\nSample DENOMINAZIONE values:")
print(categorizzazione_df['DENOMINAZIONE'].head(10).tolist())
```

```python
# Clean and standardize the key columns for better matching
fatture_df['RAGIONE_SOCIALE_CLEAN'] = fatture_df['RAGIONE_SOCIALE'].str.strip().str.upper()
categorizzazione_df['DENOMINAZIONE_CLEAN'] = categorizzazione_df['DENOMINAZIONE'].str.strip().str.upper()

# Check how many matches we have
fatture_keys = set(fatture_df['RAGIONE_SOCIALE_CLEAN'].unique())
categorizzazione_keys = set(categorizzazione_df['DENOMINAZIONE_CLEAN'].unique())

matches = fatture_keys.intersection(categorizzazione_keys)
fatture_not_matched = fatture_keys - categorizzazione_keys
categorizzazione_not_used = categorizzazione_keys - fatture_keys

print(f"MATCHING ANALYSIS:")
print(f"  Total unique companies in fatture: {len(fatture_keys)}")
print(f"  Total unique companies in categorizzazione: {len(categorizzazione_keys)}")
print(f"  Exact matches found: {len(matches)}")
print(f"  Companies in fatture NOT found in categorizzazione: {len(fatture_not_matched)}")
print(f"  Companies in categorizzazione NOT used: {len(categorizzazione_not_used)}")

# Show some examples of non-matched items
if fatture_not_matched:
    print(f"\nFirst 10 companies in FATTURE not found in categorizzazione:")
    for i, company in enumerate(list(fatture_not_matched)[:10]):
        print(f"  {i+1}. {company}")

if categorizzazione_not_used:
    print(f"\nFirst 10 companies in CATEGORIZZAZIONE not used:")
    for i, company in enumerate(list(categorizzazione_not_used)[:10]):
        print(f"  {i+1}. {company}")
```

```python
# Extract all fatture rows for companies that are NOT matched in categorizzazione
print("Extracting fatture rows for unmatched companies...")

# Filter fatture_df to get only rows where RAGIONE_SOCIALE_CLEAN is in fatture_not_matched
unmatched_fatture_rows = fatture_df[fatture_df['RAGIONE_SOCIALE_CLEAN'].isin(fatture_not_matched)]

print(f"\nUNMATCHED FATTURE ANALYSIS:")
print(f"  Total unmatched invoice rows: {len(unmatched_fatture_rows):,}")
print(f"  Unique unmatched companies: {len(fatture_not_matched)}")
print(f"  Total value of unmatched invoices: €{unmatched_fatture_rows['IMPORTO_TOTALE_DOCUMENTO'].sum():,.2f}")

# Show breakdown by company
print(f"\nBREAKDOWN BY UNMATCHED COMPANY:")
company_summary = unmatched_fatture_rows.groupby('RAGIONE_SOCIALE').agg({
    'IMPORTO_TOTALE_DOCUMENTO': ['count', 'sum']
}).round(2)

company_summary.columns = ['Invoice_Count', 'Total_Value']
company_summary = company_summary.sort_values('Total_Value', ascending=False)

print("Top 10 unmatched companies by value:")
print(company_summary.head(10).to_string())

# Show sample of unmatched invoice records
print(f"\nSAMPLE OF UNMATCHED INVOICE RECORDS:")
sample_cols = ['DATA', 'RAGIONE_SOCIALE', 'IMPORTO_TOTALE_DOCUMENTO', 'DESCRIZIONE']
available_cols = [col for col in sample_cols if col in unmatched_fatture_rows.columns]
print(unmatched_fatture_rows[available_cols].head(10).to_string(index=False))

```

```python
# RIGHT JOIN on original fatture_df but only for matched companies
print("Performing RIGHT JOIN on original fatture data for matched companies...")

# First, read the original fatture data again (in case it was modified)
original_fatture_df = pd.read_csv('fatture.csv')
original_fatture_df['RAGIONE_SOCIALE_CLEAN'] = original_fatture_df['RAGIONE_SOCIALE'].str.strip().str.upper()

# Filter categorizzazione to keep only companies that have matches in fatture
categorizzazione_matched = categorizzazione_df[categorizzazione_df['DENOMINAZIONE_CLEAN'].isin(matches)]

print(f"SETUP FOR RIGHT JOIN:")
print(f"  Original fatture records: {len(original_fatture_df):,}")
print(f"  Total categorizzazione entries: {len(categorizzazione_df):,}")
print(f"  Categorizzazione entries with matches: {len(categorizzazione_matched):,}")

# Perform RIGHT JOIN - categorizzazione_matched as the "right" table
right_joined_original = pd.merge(
    original_fatture_df,
    categorizzazione_matched,
    left_on='RAGIONE_SOCIALE_CLEAN', 
    right_on='DENOMINAZIONE_CLEAN', 
    how='left',
    suffixes=('_fatture', '_categorizzazione')
)

print(f"\nRIGHT JOIN RESULTS:")
print(f"  Records after RIGHT JOIN: {len(right_joined_original):,}")
print(f"  Unique companies in result: {right_joined_original['RAGIONE_SOCIALE'].nunique()}")

# Verify all records have both fatture and categorization data
complete_records = right_joined_original[right_joined_original['RAGIONE_SOCIALE'].notnull()]
print(f"  Complete records (with both data): {len(complete_records):,}")



# Show sample of the right joined data
print(f"\nSample of RIGHT JOINED data:")
sample_cols = ['DATA', 'RAGIONE_SOCIALE', 'IMPORTO_TOTALE_DOCUMENTO', 'MACRO_CATEGORIA', 'FISSO_VARIABILE_INVESTIMENTI', 'CATEGORIA']
available_cols = [col for col in sample_cols if col in complete_records.columns]
print(complete_records[available_cols].head(20).to_string(index=False))
```

```python
# ANALYSIS: Why RIGHT JOIN results in fewer records
print("ANALYZING WHY RIGHT JOIN HAS FEWER RECORDS...")
print("="*60)

# Let's compare the numbers step by step
print("STEP-BY-STEP ANALYSIS:")

# 1. Original fatture data
print(f"1. Original fatture records: {len(original_fatture_df):,}")

# 2. How many fatture records are for matched companies?
fatture_for_matched_companies = original_fatture_df[original_fatture_df['RAGIONE_SOCIALE_CLEAN'].isin(matches)]
print(f"2. Fatture records for MATCHED companies: {len(fatture_for_matched_companies):,}")

# 3. Categorizzazione matched entries
print(f"3. Categorizzazione matched entries: {len(categorizzazione_matched):,}")

# 4. Right join result
print(f"4. RIGHT JOIN result: {len(right_joined_original):,}")

print(f"\nEXPLANATION:")
print(f"RIGHT JOIN keeps ALL records from the RIGHT table (categorizzazione_matched)")
print(f"and matches them with the LEFT table (original_fatture_df)")

# The key insight: RIGHT JOIN result should equal the number of categorizzazione entries
# multiplied by the number of fatture records per company
print(f"\nDETAILED BREAKDOWN:")

# Check if there are multiple fatture records per company
fatture_per_company = fatture_for_matched_companies.groupby('RAGIONE_SOCIALE_CLEAN').size()
print(f"Companies with multiple invoices:")
companies_with_multiple = fatture_per_company[fatture_per_company > 1]
print(f"  Companies with >1 invoice: {len(companies_with_multiple)}")
print(f"  Max invoices per company: {fatture_per_company.max()}")
print(f"  Average invoices per company: {fatture_per_company.mean():.1f}")

# Show some examples
if len(companies_with_multiple) > 0:
    print(f"\nTop 5 companies with most invoices:")
    top_companies = fatture_per_company.nlargest(5)
    for company, count in top_companies.items():
        print(f"  {company}: {count} invoices")

# The math should be:
# RIGHT JOIN records = sum of (invoices per matched company)
expected_records = fatture_for_matched_companies.shape[0]
print(f"\nMATH CHECK:")
print(f"  Expected RIGHT JOIN records: {expected_records:,}")
print(f"  Actual RIGHT JOIN records: {len(right_joined_original):,}")
print(f"  Difference: {expected_records - len(right_joined_original):,}")

if expected_records == len(right_joined_original):
    print("✅ Numbers match perfectly! This is expected behavior.")
else:
    print("❓ There's a discrepancy. Let's investigate...")
    
    # Check for potential issues
    print(f"\nPOSSIBLE REASONS FOR DISCREPANCY:")
    
    # Check for duplicates in categorizzazione
    duplicates_in_cat = categorizzazione_matched['DENOMINAZIONE_CLEAN'].duplicated().sum()
    if duplicates_in_cat > 0:
        print(f"  - Duplicate companies in categorizzazione: {duplicates_in_cat}")
    
    # Check for null values
    null_ragione_sociale = original_fatture_df['RAGIONE_SOCIALE_CLEAN'].isnull().sum()
    if null_ragione_sociale > 0:
        print(f"  - Null RAGIONE_SOCIALE in fatture: {null_ragione_sociale}")

print(f"\n" + "="*60)
print("SUMMARY: RIGHT JOIN gives you exactly the records that can be categorized,")
print("which is typically fewer than the total original records because it excludes")
print("companies that don't exist in your categorization mapping.")
```
