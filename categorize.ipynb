```python
import pandas as pd
import numpy as np
import os
import glob
import shutil
from difflib import SequenceMatcher

```

```python


# Create directories for organization
parquet_dir = "parquet_files"
csv_dir = "csv_files"

os.makedirs(parquet_dir, exist_ok=True)
os.makedirs(csv_dir, exist_ok=True)

# Find all files with parquet extention in the current directory
parquet_files = glob.glob("*.parquet")
print(f"Found {len(parquet_files)} parquet files:")
for file in parquet_files:
    print(f"  - {file}")



# Process each parquet file
for parquet_file in parquet_files:
    print(f"\nProcessing: {parquet_file}")
    
    # Read the parquet file
    df = pd.read_parquet(parquet_file)
    print(f"  Shape: {df.shape}")
    
    # Convert to CSV and save in csv_files directory
    csv_filename = parquet_file.replace('.parquet', '.csv')
    csv_path = os.path.join(csv_dir, csv_filename)
    df.to_csv(csv_path, index=False)
    print(f"  Saved CSV as: {csv_path}")
    
    # Move parquet file to parquet_files directory
    parquet_path = os.path.join(parquet_dir, parquet_file)
    shutil.move(parquet_file, parquet_path)
    print(f"  Moved parquet to: {parquet_path}")
    
    # Add to list for combining


print(f"\n{'='*50}")
print("SUMMARY:")
print(f"Individual CSV files created in '{csv_dir}/' directory: {len(parquet_files)}")
print(f"Parquet files moved to '{parquet_dir}/' directory: {len(parquet_files)}")
```

```python

# Fetch all CSV files under the 'csv' directory
csv_files = glob.glob('csv/*.csv')

# Read each CSV file into a DataFrame and store in all_dataframes
all_dataframes = [pd.read_csv(f) for f in csv_files]

print(f"Loaded {len(all_dataframes)} CSV files from 'csv/' directory.")
```

```python
# Combine all dataframes into one
if all_dataframes:
    combined_df = pd.concat(all_dataframes, ignore_index=True)
    
    print(f"Combined dataset shape: {combined_df.shape}")
    print(f"Total rows: {combined_df.shape[0]:,}")
    print(f"Total columns: {combined_df.shape[1]}")
    
    # Save combined data to CSV in the csv_files directory
    combined_csv_filename = "all_data_combined.csv"
    combined_csv_path = os.path.join(csv_dir, combined_csv_filename)
    combined_df.to_csv(combined_csv_path, index=False)
    print(f"\nCombined data saved as: {combined_csv_path}")
    
    # Display basic info about the combined dataset
    print(f"\nColumns in combined dataset:")
    for i, col in enumerate(combined_df.columns, 1):
        print(f"  {i}. {col}")
    
    print(f"\nFirst few rows of combined data:")
    print(combined_df.head(3))
    
    # Check for any data quality issues
    print(f"\nData Quality Check:")
    print(f"  Missing values per column:")
    missing_values = combined_df.isnull().sum()
    for col, missing in missing_values.items():
        if missing > 0:
            print(f"    {col}: {missing} missing values")
    
    if missing_values.sum() == 0:
        print("    No missing values found!")
        
    # Final directory structure summary
    print(f"\n{'='*60}")
    print("FINAL ORGANIZATION:")
    print(f"📁 {csv_dir}/ - Contains all CSV files (individual + combined)")
    print(f"   └── Combined CSV file: {combined_csv_filename}")
else:
    print("No dataframes to combine!")
```

```python
# Clean and standardize the key columns for better matching
fatture_df = fatture_df_origin.copy()
categorizzazione_df = categorizzazione_df_origin.copy()
fatture_df['RAGIONE_SOCIALE_CLEAN'] = fatture_df['RAGIONE_SOCIALE'].str.strip().str.upper()
categorizzazione_df['DENOMINAZIONE_CLEAN'] = categorizzazione_df['DENOMINAZIONE'].str.strip().str.upper()

# Check how many matches we have
fatture_ragione_sociale = set(fatture_df['RAGIONE_SOCIALE_CLEAN'].unique())
categorizzazione_denominazione = set(categorizzazione_df['DENOMINAZIONE_CLEAN'].unique())

matches = fatture_ragione_sociale.intersection(categorizzazione_denominazione)
fatture_not_matched = fatture_ragione_sociale - categorizzazione_denominazione
categorizzazione_not_used = categorizzazione_denominazione - fatture_ragione_sociale

print(f"MATCHING ANALYSIS:")
print(f"  Total unique companies in fatture: {len(fatture_ragione_sociale)}")
print(f"  Total unique companies in categorizzazione: {len(categorizzazione_denominazione)}")
print(f"  Exact matches found: {len(matches)}")
print(f"  Companies in fatture NOT found in categorizzazione: {len(fatture_not_matched)}")
print(f"  Companies in categorizzazione NOT used: {len(categorizzazione_not_used)}")

# Show some examples of non-matched items
if fatture_not_matched:
    print(f"\nFirst 10 companies in FATTURE not found in categorizzazione:")
    for i, company in enumerate(list(fatture_not_matched)[:10]):
        print(f"  {i+1}. {company}")
        # Print all fatture records for this company
        company_fatture = fatture_df[fatture_df['RAGIONE_SOCIALE_CLEAN'] == company]
        print(f"     Found {len(company_fatture)} invoice(s) for this company:")
        print(company_fatture.to_string(index=False))
        print("-" * 80)

if categorizzazione_not_used:
    print(f"\nFirst 10 companies in CATEGORIZZAZIONE not used:")
    for i, company in enumerate(list(categorizzazione_not_used)[:10]):
        print(f"  {i+1}. {company}")
```

```python
# Check if all P_IVA values in fatture_df are present in categorizzazione_df and vice versa
fatture_piva_set = set(fatture_df['P_IVA'].dropna().astype(str))
categorizzazione_piva_set = set(categorizzazione_df['P_IVA'].dropna().astype(str))

piva_matches = fatture_piva_set.intersection(categorizzazione_piva_set)
fatture_piva_not_matched = fatture_piva_set - categorizzazione_piva_set
categorizzazione_piva_not_matched = categorizzazione_piva_set - fatture_piva_set

print(f"P_IVA MATCHING ANALYSIS:")
print(f"  Unique P_IVA in fatture: {len(fatture_piva_set)}")
print(f"  Unique P_IVA in categorizzazione: {len(categorizzazione_piva_set)}")
print(f"  Matches found: {len(piva_matches)}")
print(f"  P_IVA in fatture NOT in categorizzazione: {len(fatture_piva_not_matched)}")
print(f"  P_IVA in categorizzazione NOT in fatture: {len(categorizzazione_piva_not_matched)}")

if fatture_piva_not_matched:
    print("\nSample P_IVA in fatture not matched in categorizzazione:")
    print(list(fatture_piva_not_matched)[:10])

if categorizzazione_piva_not_matched:
    print("\nSample P_IVA in categorizzazione not matched in fatture:")
    print(list(categorizzazione_piva_not_matched)[:10])
```

```python
from difflib import SequenceMatcher

# Prepare sets of P_IVA as strings for comparison
fatture_piva_list = sorted(set(str(p) for p in fatture_df['P_IVA'].dropna()))
categorizzazione_piva_list = sorted(set(str(p) for p in categorizzazione_df['P_IVA'].dropna()))

similarity_results = []

for piva in fatture_piva_list:
    # Find the most similar P_IVA in categorizzazione
    best_match = None
    best_score = 0
    for cat_piva in categorizzazione_piva_list:
        score = SequenceMatcher(None, piva, cat_piva).ratio()
        if score > best_score:
            best_score = score
            best_match = cat_piva
    # Only consider if not an exact match
    if best_match and piva != best_match:
        # Get a sample ragione sociale for this P_IVA
        ragione_sociale = fatture_df.loc[fatture_df['P_IVA'].astype(str) == piva, 'RAGIONE_SOCIALE'].iloc[0]

        # Get categoria from categorizzazione for the best matching P_IVA
        categoria = categorizzazione_df.loc[categorizzazione_df['P_IVA'].astype(str) == best_match, 'CATEGORIA'].iloc[0]

        similarity_results.append((ragione_sociale, piva, best_match, best_score, categoria))

# Sort by similarity descending and show 
similarity_results_sorted = sorted(similarity_results, key=lambda x: -x[3])

# Display as DataFrame for clarity
similarity_df = pd.DataFrame(similarity_results_sorted, columns=['RAGIONE_SOCIALE', 'P_IVA', 'P_IVA_fornitore', 'similarity','Categoria'])
print(similarity_df.head(100).to_string(index=False))
```

```python
# List of all different DataFrames created so far in this notebook:

# 1. df: Used in the initial parquet/csv conversion and combination process (not shown directly, but implied).
# 2. all_dataframes: List of DataFrames loaded from individual CSV files.
# 3. combined_df: The concatenated DataFrame of all CSVs.
# 4. fatture_df: DataFrame loaded from 'fatture.csv', containing invoice data.
# 5. categorizzazione_df: DataFrame loaded from 'categorizzazione_fornitori.csv', containing supplier categorization.
# 6. original_fatture_df: A fresh read of 'fatture.csv' for join operations.
# 7. similarity_df: DataFrame showing similarity matches between P_IVA values.
```

```python
import pandas as pd
from difflib import SequenceMatcher

# Create a mapping dictionary for P_IVA corrections based on similarity analysis
piva_mapping = {}

# Add exact matches
for piva in piva_matches:
    piva_mapping[piva] = piva

# Add similar matches from similarity analysis
for _, row in similarity_df.iterrows():
    fatture_piva = row['P_IVA']
    categorizzazione_piva = row['P_IVA_fornitore']
    piva_mapping[fatture_piva] = categorizzazione_piva

# Create a mapped P_IVA column in fatture_df for joining
fatture_df['P_IVA_MAPPED'] = fatture_df['P_IVA'].astype(str).map(piva_mapping)

# Fill unmapped values with original P_IVA (for exact matches)
fatture_df['P_IVA_MAPPED'] = fatture_df['P_IVA_MAPPED'].fillna(fatture_df['P_IVA'].astype(str))

# Perform the join
merged_df = fatture_df.merge(
    categorizzazione_df, 
    left_on='P_IVA_MAPPED', 
    right_on='P_IVA', 
    how='left',
    suffixes=('_fatture', '_categorizzazione')
)

print(f"Original fatture shape: {fatture_df.shape}")
print(f"Merged dataframe shape: {merged_df.shape}")
print(f"Successfully matched records: {merged_df['CATEGORIA'].notna().sum()}")
print(f"Unmatched records: {merged_df['CATEGORIA'].isna().sum()}")

print(merged_df.sample(20).to_string(index=False)) 
# Show sample of merged data
print("\nSample of merged data:")
print(merged_df[['RAGIONE_SOCIALE', 'P_IVA_fatture', 'P_IVA_categorizzazione', 'CATEGORIA', 'MACRO_CATEGORIA']].head(10))
```

```python
merged_df.rename(columns={
    'P_IVA_MAPPED': 'P_IVA'

}, inplace=True)

# Drop the original RAGIONE_SOCIALE column and rename DENOMINAZIONE to RAGIONE_SOCIALE
merged_df = merged_df.drop(columns=['RAGIONE_SOCIALE'])
merged_df = merged_df.rename(columns={'DENOMINAZIONE': 'RAGIONE_SOCIALE'})



colunmns_to_keep = fatture_df_origin.columns.tolist() + ['CATEGORIA', 'MACRO_CATEGORIA']
fatture_df_final = merged_df[colunmns_to_keep].copy()

# Format CODICE_COMMITTENTE as 11-digit integer with leading zeros, remove .0 suffix
fatture_df_final['CODICE_COMMITTENTE'] = fatture_df_final['CODICE_COMMITTENTE'].astype(str).str.replace('.0', '', regex=False).str.zfill(11)
# Format ALIQUOTA_IVA as integer, remove .0 suffix
fatture_df_final['ALIQUOTA_IVA'] = fatture_df_final['ALIQUOTA_IVA'].astype(str).str.replace('.0', '', regex=False).astype(int)
# Add enumeration column as primary key at the first position
fatture_df_final.insert(0, 'ID', range(1, len(fatture_df_final) + 1))

print(fatture_df_final.shape)
print(fatture_df_final.sample(20).to_string(index=False))
fatture_df_final.to_csv('fatture_categorizzate.csv', index=False)

```

```python
import pandas as pd
import os
import glob

# Directory containing parquet files
parquet_dir = "parquet"
csv_dir = "csv_files"
output_csv = "combined_data.csv"

# Ensure output directory exists
os.makedirs(csv_dir, exist_ok=True)

# Find all parquet files in the directory
parquet_files = glob.glob(os.path.join(parquet_dir, "*.parquet"))
print(f"Found {len(parquet_files)} parquet files.")

# List to store DataFrames
dataframes = []

# Process each parquet file
for parquet_file in parquet_files:
    print(f"Processing: {parquet_file}")
    df = pd.read_parquet(parquet_file)
    dataframes.append(df)

   

# Concatenate all DataFrames
if dataframes:
    combined_df = pd.concat(dataframes, ignore_index=True)
    combined_df.to_csv(output_csv, index=False)
    print(f"Combined data saved to: {output_csv}")
else:
    print("No parquet files to process.")
```
